````python
import socket
import json

# 创建一个 TCP 套接字
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 服务器地址和端口
server_address = ('114.94.100.133', 8888)

try:
    # 连接到服务器
    client_socket.connect(server_address)
    # 构造一个示例请求
    request = {"message": "这是一个测试请求"}
    # 将请求转换为 JSON 字符串并编码为字节
    request_data = json.dumps(request).encode('utf-8')
    # 发送请求
    client_socket.sendall(request_data)
    # 接收服务器的确认消息
    confirmation = client_socket.recv(1024)
    print(f"服务器确认: {confirmation.decode('utf-8')}")
    # 接收服务器的响应
    response = client_socket.recv(1024)
    print(f"服务器响应: {response.decode('utf-8')}")
except Exception as e:
    print(f"连接出错: {e}")
finally:
    # 关闭连接
    client_socket.close()
````

````python
import re
def remove_markdown(text):
    text = re.sub(r'\*\*(.*?)\*\*', r'\1', text)  # 移除粗体 **文本**
    text = re.sub(r'\*(.*?)\*', r'\1', text)      # 移除斜体 *文本*
    
    # 移除标题标记 (# 标题)
    text = re.sub(r'^#+\s+', '', text, flags=re.MULTILINE)
    
    # 移除有序列表标记 (1. 列表项)，包括缩进的列表
    text = re.sub(r'^\s*\d+\.\s+', '', text, flags=re.MULTILINE)
    
    # 移除无序列表标记 (- 列表项 或 * 列表项)，包括缩进的列表
    text = re.sub(r'^\s*[\-\*]\s+', '', text, flags=re.MULTILINE)
    
    # 移除链接，保留链接文本 ([链接文本](链接URL))
    text = re.sub(r'\[(.*?)\]\(.*?\)', r'\1', text)
    
    # 移除图片标记 (![替代文本](图片URL))
    text = re.sub(r'!\[.*?\]\(.*?\)', '', text)
    
    # 移除代码块
    text = re.sub(r'```.*?```', '', text, flags=re.DOTALL)
    
    # 移除行内代码标记
    text = re.sub(r'`(.*?)`', r'\1', text)
    
    # 移除引用标记
    text = re.sub(r'^>\s+', '', text, flags=re.MULTILINE)
    
    # 移除水平分割线
    text = re.sub(r'^\s*[\-\*_]{3,}\s*$', '', text, flags=re.MULTILINE)

    return text.strip()

text = "- **分类法开发与验证**：通过多轮注释者间的一致性研究，开发并验证了多智能体系统故障分类法（MASFT），同时还开发了基于LLM的注释器来自动分析MAS跟踪。\
3. **研究发现**\
    - **失败类别**：确定了三个主要的失败类别，包括规范和系统设计失败、智能体间失调以及任务验证和终止。\
    - **失败模式**：共识别出14种细粒度的失败模式，例如违反任务规范、角色规范不遵守、步骤重复、对话历史丢失等。\
    - **相关性**：不同的失败类别和模式之间存在一定的相关性，且不同的MAS表现出不同的失败分布。\
4. **案例研究**\
    - **AG2 - MathChat**：通过改进提示和优化智能体配置，提高了性能，但效果因底层LLM而异。"

text = remove_markdown(text)
print(text)
````
